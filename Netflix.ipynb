# Task 4

```python
# to fill Nan to zero
dd=dd.fillna(0)  
# get year from Release Date
dd['Year'] = pd.to_datetime(dd['Release_Date'].str.strip(), format="%B %d, %Y").dt.year 
# convert Year into integer
dd['Year'] = dd['Year'].astype('Int64')   
dd
```

```python
# filter Tv Show from Category
filter_TvShow=dd[dd['Category'] == 'TV Show']
# filter Movie from Category
filter_Movie=dd[dd['Category'] == 'Movie']
```

```python
# to get count of 'Tv Show' respective year , groupby year
TVShow_Count = filter_TvShow.groupby('Year').size().reset_index(name='TVShow_Count')
print('Number of TV Show released Year wise')
TVShow_Count
```

```python
# bar plot No of Tv Show Released By Year
plt.figure(figsize=(10,6)) # x,y
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.bar(TVShow_Count['Year'], TVShow_Count['TVShow_Count'], color=colors, edgecolor='black')
plt.xlabel('Released_Year')
plt.ylabel('Tv Show Count')
plt.title('No of Tv Show Released By Year')
```

```python
# to get count of 'Movie' released respective year , groupby year
Movie_count = filter_Movie.groupby('Year').size().reset_index(name='Movie_Count')
print('Number of Movies released Year wise')
Movie_count
```

```python
# bar plot for No of Movies Released By Year
plt.figure(figsize=(10,6)) # x,y
plt.bar(Movie_count['Year'], Movie_count['Movie_Count'], color=colors, edgecolor='black')
plt.xlabel('Released_Year')
plt.ylabel('Movie Count')
plt.title('No of Movies Released By Year')
```

```python
# maximum and minimum duration of a movie
movies=dd[dd['Duration'].str.contains('min',case=False,na=False)]
movies = movies.copy()
movies['minutes'] =movies['Duration'].str.extract(r'(\d+)').astype('int')
max_movie_duration=movies['minutes'].max()
min_movie_duration=movies['minutes'].min()
print(f"Maximum movie duration: {max_movie_duration} minutes")
print(f"Minimum movie duration: {min_movie_duration} minutes")
```

```python
# maximum and minimum duration of a Tv show
tv_shows=dd[dd['Duration'].str.contains('Season',case=False,na=False)]
tv_shows = tv_shows.copy()
tv_shows['Season'] =tv_shows['Duration'].str.extract(r'(\d+)').astype('int')
max_tvshow_duration=tv_shows['Season'].max()
min_tvshow_duration=tv_shows['Season'].min()
print(f"Maximum TV show duration: {max_tvshow_duration} seasons")
print(f"Minimum TV show duration: {min_tvshow_duration} seasons")
```

# Task 9

```python
# Create a copy of the original DataFrame due to some error
dd_copy = dd.copy()

# Country has the highest no. of.Tv Shows

# Split the 'Country' column into lists  (split the multiple country in the single column based on comma)
dd_copy['Country'] = dd_copy['Country'].str.split(',')

# Explode the 'Country' column to flatten the lists (list the county in seperate rows)
dd_copy = dd_copy.explode('Country')

# Strip leading/trailing spaces from 'Country' values
dd_copy['Country'] = dd_copy['Country'].str.strip()
dd_copy
```

```python
# get no .of tv shows for each country
grp_country=dd_copy.groupby('Country').agg(TV_Show=('Category',lambda x:(x=='TV Show').sum()))
```

```python
grp_country
```

```python
# plot for no .of tv shows release by eash country
plt.figure(figsize=(20,10)) # x,y
plt.bar(grp_country.index,grp_country['TV_Show'],color='orange',edgecolor='red')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# maximum tvshow released country
max_row = grp_country[grp_country['TV_Show'] == grp_country['TV_Show'].max()]
max_row
```

```python
# Movie count for each country
grp_countr_movie=dd_copy.groupby('Country').agg(Movie=('Category',lambda x:(x=='Movie').sum()))
grp_countr_movie
```

```python
# Comparison of TV Show and Movie Counts by Country
plt.figure(figsize=(20,10)) # x,y
x = np.arange(len(grp_country))
y = np.arange(len(grp_countr_movie))

# Create the grouped bar plot
plt.bar(x - 0.2, grp_country['TV_Show'], width=0.4, label='TV Show', color='red')
plt.bar(y + 0.2, grp_countr_movie['Movie'], width=0.4, label='Movie', color='blue')

# Add title and labels
plt.title('Comparison of TV Show and Movie Counts by Country')
plt.xlabel('Country')
plt.ylabel('Count')


# Set x-axis ticks to correspond to country names
plt.xticks(x, grp_country.index, rotation=90)

# Add a legend
plt.legend()

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# maximum movie released by country
max_row_movie = grp_countr_movie[grp_countr_movie['Movie'] == grp_countr_movie['Movie'].max()]
max_row_movie
```
